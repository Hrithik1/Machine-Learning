# Decision Trees

```python
import numpy as np
import pandas as pd
from sklearn import tree

file="/Users/hrithikrana/Documents/GitHub/Machine-Learning/PastHires.csv"
data=pd.read_csv(file,header=0)
```

```python
data.head()
```

```python
d={'Y':1,'N':0}
data['Employed?']=data['Employed?'].map(d)
data['Top-tier school']=data['Top-tier school'].map(d)
data['Interned']=data['Interned'].map(d)
data['Hired']=data['Hired'].map(d)
e={'BS':0,'MS':1,'PhD':2}
data['Level of Education']=data['Level of Education'].map(e)
```

```python
data.head()
```

```python
features=list(data.columns[:6])
features
```

```python
y=data['Hired']
z=data[features]
dec_tree=tree.DecisionTreeClassifier()
dec_tree=dec_tree.fit(z,y)
```
