# Recommender System

Item based collaborative filtering

```python
import pandas as pd

r_cols=['user_id','movie_id','rating']
ratings=pd.read_csv("/Users/hrithikrana/Downloads/MLCourse/Machinelearning-pandas-numpy-scipy-/ml-100k/u.data",sep='\t',names=r_cols,usecols=range(3), encoding="ISO-8859-1")

m_cols=['movie_id','title']
movies=pd.read_csv("/Users/hrithikrana/Downloads/MLCourse/Machinelearning-pandas-numpy-scipy-/ml-100k/u.item",sep='|',names=m_cols,usecols=range(2), encoding="ISO-8859-1")

ratings=pd.merge(movies,ratings)
```

```python
ratings.head()
```

```python
movie_rating=ratings.pivot_table(index='user_id',columns='title',values='rating')
movie_rating.head()
```

```python
star_wars_rating=movie_rating['Star Wars (1977)']
star_wars_rating.head()
```

```python
similar_movies=movie_rating.corrwith(star_wars_rating)
similar_movies=similar_movies.dropna()
df=pd.DataFrame(similar_movies)
df.head(10)
```

```python
similar_movies.sort_values(ascending=False)
```

```python
import numpy as np
movie_stats=ratings.groupby('title').agg({'rating':[np.size,np.mean]})
movie_stats.head(10)
```

```python
popular_movies=movie_stats['rating']['size']>=300
movie_stats[popular_movies].sort_values([('rating','mean')],ascending=False)[:20]
```

```python
df=movie_stats[popular_movies].join(pd.DataFrame(similar_movies,columns=['similarity']))
df.head(20)
```

```python
df.sort_values((['similarity']),ascending=False)[:20]
```
