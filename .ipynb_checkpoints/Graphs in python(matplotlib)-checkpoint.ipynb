# Types of Graphs

1) Line Graph

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
x=np.arange(-15,15,.001)
plt.plot(x,norm.pdf(x))
plt.show()
```

2) Multiple plots on one graph and Save to a file

```python
plt.plot(x,norm.pdf(x,1,0.5))
plt.plot(x,norm.pdf(x,5,0.5))
plt.plot(x,norm.pdf(x))
plt.savefig("/Users/hrithikrana/Desktop/Multipleplot.png",format="png")
```

3) Grid Lines


```python
a=plt.axes()
a.grid()
plt.plot(x,norm.pdf(x,5,.5))
```

4) Colour and Line type

```python
from scipy.stats import norm
x=np.arange(-10,10,0.5)
plt.plot(x,norm.pdf(x,1),'r-')
plt.plot(x,norm.pdf(x,2),'g:')
plt.plot(x,norm.pdf(x,3),'b--')
plt.xlabel("A")
plt.ylabel("B")
plt.legend(["Red","Green","Blue"],loc=1)
a.grid()
plt.savefig("/Users/hrithikrana/Desktop/RGBgraph.png",format='png')
```

5)  Pie Chart

```python
values=[10,20,30,40,50,60]
colors=['r','g','b','y','c','m']
explode=[0,0,0,0,0,0.2]
labels=['USA','China','Europe','Australia','Japan','India']
plt.pie(values,colors=colors,labels=labels,explode=explode)
plt.title("Data")
plt.show()
```
