# Types of Graphs

1) Line Graph

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
x=np.arange(-15,15,.001)
plt.plot(x,norm.pdf(x))
plt.show()
```

2) Multiple plots on one graph and Save to a file

```python
plt.plot(x,norm.pdf(x,1,0.5))
plt.plot(x,norm.pdf(x,5,0.5))
plt.plot(x,norm.pdf(x))
plt.savefig("/Users/hrithikrana/Desktop/Multipleplot.png",format="png")
```
